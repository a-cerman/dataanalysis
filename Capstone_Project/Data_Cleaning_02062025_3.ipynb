```python
def clean_item(x):
    x = str(x)
    x = re.sub(r'[\n\t\r]', '', x)  # Remove control characters

    # Allowed meaningful characters
    allowed = 'a-zA-Z0-9€₺$%-+./'

    # Remove unwanted chars at the start
    x = re.sub(fr'^[^{allowed}]+', '', x)

    # Remove unwanted chars at the end
    x = re.sub(fr'[^{allowed}]+$', '', x)

    return x.strip()
```

```python
def clean_cell(value):
    # return missing/null values as np.nan
    if not isinstance(value, list) and pd.isna(value):
        return np.nan

    # Helper to clean each item: remove \n, \t, \r and strip
    #def clean_item(x):
    #    return re.sub(r'[\n\t\r]', ',', str(x)).str.replace(r'[()]', ' ',regex=True)strip()
        
    if isinstance(value, list):
        # Flatten nested list and clean each element
        flat = []
        for item in value:
            if isinstance(item, list):
                flat.extend(item)
            else:
                flat.append(item)
        # Clean each flattened item
        #return ';'.join(str(x).strip() for x in flat if x is not None)
        #return tuple(clean_item(x) for x in flat if x is not None)
        cleaned = [clean_item(x) for x in flat if x is not None and str(x).strip() != '']

        if len(cleaned) == 0:
            return np.nan  # Optional: treat empty result as NaN
        elif len(cleaned) == 1:
            return cleaned[0]  # Just a single string
        else:
            return tuple(cleaned)  # A tuple if multiple items
            
    # If it's just a string, clean it    
    elif isinstance(value, str):
        # Remove unwanted whitespace characters and trailing spaces
         return clean_item(value)
        
    # Leave numbers or other types unchanged 
    else:
        return value
```

```python
# def clean_price(price):
#     try:
#         price = re.sub(r"[^\d,]", "", str(price))
#         return float(price.replace(",", "")) if price else None
#     except:
#         return None

def clean_price_column(df, column_name):
    return df[column_name] \
             .astype(str) \
             .str.replace(r"[^\d,]", "", regex=True) \
             .str.replace(",", "") \
             .astype(float)

def extract_country(location):
    try:
        parts = str(location).strip().split()
        return parts[-1] if parts else None
    except:
        return None

def clean_warranty(warranty):
    try:
        match = re.search(r'(\d+)', str(warranty))
        return int(match.group(1)) if match else None
    except:
        return None

# def clean_mileage(mileage):
#     try:
#         mileage = re.sub(r"[^\d]", "", str(mileage))
#         return int(mileage) if mileage else None
#     except:
#         return None

def clean_mileage(mileage):
    try:
        match = re.search(r"([\d.,]+)", str(mileage))
        if match:
            number = match.group(1)
            number = number.replace(",", "")  # "120,200" → "120200"
            return float(number)
        else:
            return None
    except:
        return None


def extract_horsepower(power):
    try:
        match = re.search(r'\((\d+)\s*hp\)', str(power))
        return int(match.group(1)) if match else None
    except:
        return None

def extract_engine_size(size):
    try:
        # Stringe çevir, yalnızca sayı ve virgül olan kısmı ayıkla
        match = re.search(r"([\d.,]+)", str(size))
        if match:
            # Bulunan sayı kısmını al
            number = match.group(1)
            # Virgülü kaldır ve noktaya çevirme: "1,991" → "1991"
            number = number.replace(",", "")
            return float(number)
        else:
            return None
    except:
        return None

# empty_weight sütunu
def clean_numeric_column_inplace(df, column_name):
    """
    DataFrame içindeki bir sütundaki '1,395 kg' gibi değerlerden sadece sayısal kısmı alır
    ve aynı sütunu float olarak günceller.
    """
    df[column_name] = (
        df[column_name]
          .astype(str)
          .str.extract(r'([\d.,]+)')[0]   # sadece sayıyı al
          .str.replace(",", "")           # virgülü kaldır
          .astype(float)                  # float'a çevir
    )

# # empty_weight sütunu
# def clean_weight(val):
#     if pd.isna(val):
#         return None  # NaN'ler 0 olarak döner
    
#     try:
#         # String'e çevir, "kg" çıkar, virgül sil, boşlukları temizle
#         val_str = str(val).replace('kg', '').replace(',', '').strip()
#         return int(val_str)
#     except:
#         return None
# ri1['empty_weight'] = ri1['empty_weight'].apply(clean_weight)
```

```python
# Task 10 previous_owner
def extract_user_count(val):
    # Eğer NaN ise None döndür
    if pd.isna(val):
        return 0
    
    # Eğer tuple ise son elemanı al
    if isinstance(val, tuple) or isinstance(val, list):
        try:
            return int(val[-1])
        except:
            return 0
    
    # Eğer string ise
    if isinstance(val, str):
        try:
            # Parantezleri temizle, ardından son virgülden sonrası alınır
            val_clean = val.strip("() ")
            parts = val_clean.split(",")
            return int(parts[-1].strip())
        except:
            return 0
    
    # Diğer durumlarda None
    return 0
```

```python
def clean_co2_emissions_column(df, column_name):
    """
    'g/km (comb.)' gibi metinli CO2 verilerini temizleyip sadece sayısal değeri float olarak bırakır.
    Orijinal sütun inplace olarak güncellenir.
    
    Örn: "119 g/km (comb.)" → 119.0
    """
    df[column_name] = (
        df[column_name]
          .astype(str)
          .str.extract(r'(\d+\.?\d*)')[0]   # Sayı kısmını çek
          .astype(float)                   # Float'a çevir
    )
```

```python
def map_fuel_type(fuel):
    # Check if it's not a string or it's empty
    if not isinstance(fuel, str) or fuel.strip() == '':
        return fuel
    f = fuel.lower()
    if 'others' in f:
        value= 'Others'
    if 'electric' in f:
        value='Electric'
    if 'diesel' in f:
        value = 'Diesel'
    # Petrol keywords
    if any(kw in f for kw in ['gasoline', 'super', 'benzine', 'regular', 'ethanol']):
        value ='Petrol' 
    if any(kw in f for kw in ['cng', 'domestic gas','liquid petroleum gas','biogas', 'lpg','hydrogen']):
        value = 'Gas'
    return value
```

```python
import ast
```

```python
# String olarak görünen tuple verileri güvenli biçimde dönüştür
def parse_fuel_string(value):
    if isinstance(value, str) and value.startswith("("):
        try:
            return ast.literal_eval(value)
        except (ValueError, SyntaxError):
            return None
    return value
```

```python
#fuel_consumption sütununu tuple olarak parse et
ri16['fuel_consumption'] = ri16['fuel_consumption'].apply(parse_fuel_string)
```

```python
# Ondalık veya tam sayıları string içinden çeken fonksiyon
def extract_first_decimal(text):
    if not isinstance(text, str):
        return text
    match = re.search(r'\d+\.*\d*', text.replace(',', '.'))
    return float(match.group()) if match else np.NaN
```

```python
# Tuple/str verisini yakıt türlerine göre doğru kolonlara ayırma
def split_fuel_values(value):
    result = {'fuel_comb': None, 'fuel_city': None, 'fuel_country': None}

    # Eğer liste veya tuple ise tek tek dolaş
    if isinstance(value, (list, tuple)):
        items = value
    # Eğer sadece string ise tek elemanlı listeye çevir
    elif isinstance(value, str):
        items = [value]
    else:
        return pd.Series(result)  # None, boş, yanlış tip → boş döndür

    for item in items:
        if isinstance(item, str):
            lowered = item.lower()
            if 'comb' in lowered:
                result['fuel_comb'] = extract_first_decimal(item)
            elif 'city' in lowered:
                result['fuel_city'] = extract_first_decimal(item)
            elif 'country' in lowered:
                result['fuel_country'] = extract_first_decimal(item)
       
    return pd.Series(result)
```

```python
# Decimal değerleri fuel_comb, fuel_city, fuel_country olarak ayır
ri16[['fuel_comb', 'fuel_city', 'fuel_country']] = ri16['fuel_consumption'].apply(split_fuel_values)
```
