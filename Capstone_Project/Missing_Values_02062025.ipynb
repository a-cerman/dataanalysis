```python
def tahmin_et(row):                # NAN degerleri doldurmak icin diger column'daki bilgilere gore kac ceker oldugu bulunuyor.
    model = str(row['make_model']).lower()
    body = row['body_type']
    power =  str(row['power'])#str(row['power_original'])
    
    # SUV'ler genelde 4 çeker
    if body == 'SUV':
        return 4
    
    # Model isminde GLC, GLE, 4MATIC, 4x4 varsa 4 çeker
    if any(x in model for x in ['glc', 'gle', '4matic', '4x4']):
        return 4
    
    # CITY tipi ve düşük güçlü ise 2 çeker
    if body == 'CITY' and ('kW (1' in power or '80 kW' in power or '66 kW' in power):
        return 2
    
    # Emin olamıyorsak NaN bırak
    return None

# Sadece eksik olan drivetrain satırlarına uygula
ri13.loc[ri13['drivetrain'].isna(), 'drivetrain'] = ri13[ri13['drivetrain'].isna()].apply(tahmin_et, axis=1)

# Tahmin edilenleri görelim
tahmin_edilenler = ri13[(ri13['drivetrain_original'].isna()) & (ri13['drivetrain'].notna())][['make_model', 'body_type', 'power', 'drivetrain']]
print(tahmin_edilenler.head(10))
```

```python
# task 5
# ri1.upholstery    >>>>>> ri1.upholstery_imputed


"""
upholstery_imputed

Bu fonksiyon, aracın iç döşemesi (upholstery) sütunundaki eksik değerleri 
çok aşamalı bir yaklaşımla doldurur ve ardından Premium/Standart olarak 
ikili sınıflandırma gerçekleştirir.


Adım 1: Orijinal 'upholstery' sütununu koruyarak 'upholstery_imputed' adlı yeni bir sütun oluşturur.
Adım 2: 'make_model' grubunda en az 20 kaydı olanlar için mod (en sık değer) imputasyonu uygular.
Adım 3: Aynı 'make_model' grubu içindeki eksikleri ardışık komşu doldurma (ffill + bfill) ile tamamlar.
----
Adım 4: Dolgu öncesi ve sonrası eksik sayılarını hesaplar.
Adım 5: 'full leather', 'part leather' ve 'alcantara' tiplerini Premium; diğer tüm tipleri Standart olarak etiketler.
Adım 6: Hâlâ eksik kalan satırlar için bir bayrak (upholstery_missing) ekler.
"""

# 1. Başlangıç: yeni sütunu orijinal upholstery ile oluştur
ri1['upholstery_imputed'] = ri1['upholstery']

# 2. Make–model bazlı mod imputasyon (sadece ≥20 kaydı olan gruplar)
counts = ri1['make_model'].value_counts()
eligible_models = counts[counts >= 20].index

mode_map = (
    ri1[ri1['make_model'].isin(eligible_models)]
 
     .groupby('make_model')['upholstery']
      .agg(lambda s: s.mode().iat[0] if not s.mode().empty else np.nan)
)

mask_mode = ri1['upholstery_imputed'].isna() & ri1['make_model'].isin(eligible_models)
ri1.loc[mask_mode, 'upholstery_imputed'] = ri1.loc[mask_mode, 'make_model'].map(mode_map)

# 3. Grup içi forward/backward fill (ffill + bfill)
ri1 = ri1.sort_values('make_model')

ri1['upholstery_imputed'] = (
    ri1
    .groupby('make_model')['upholstery_imputed']
    .transform(lambda s: s.ffill().bfill())
)

# 4. Kalan NaN sayısını kontrol
#remaining_nans = ri1['upholstery_imputed'].isna().sum()
#result = pd.DataFrame({
#    'count_before': ri1['upholstery'].isna().sum(),
#    'count_after': ri1['upholstery_imputed'].isna().sum()
#}, index=['upholstery'])
#print(result)


```

```python
# 0. Preparations
# ----------------
# Clean CO2 emissions
ri1['co2_emissions_clean'] = (
    ri1['CO2_emissions'].astype(str)
                      .str.extract(r'(\d+)', expand=False)
)
ri1['co2_emissions_clean'] = pd.to_numeric(ri1['co2_emissions_clean'], errors='coerce')

# Parse engine size (cc)
ri1['engine_cc'] = (
    ri1['engine_size'].astype(str)
                      .str.replace(r'[^0-9]', '', regex=True)
                      .replace('', pd.NA)
)
ri1['engine_cc'] = pd.to_numeric(ri1['engine_cc'], errors='coerce')

# Define engine_cc bins (200 cc intervals)
max_cc = int(ri1['engine_cc'].max(skipna=True))
bins = np.arange(0, max_cc + 200, 200)
labels = [f"{bins[i]}-{bins[i+1]-1}" for i in range(len(bins)-1)]
ri1['engine_cc_bin'] = pd.cut(ri1['engine_cc'], bins=bins, labels=labels, right=False)

# 1. Initialize imputed column
# -----------------------------
ri1['co2_imputed'] = ri1['co2_emissions_clean']
print("After init, NaNs:", ri1['co2_imputed'].isna().sum())

# 2. Quadruple group mode imputation
# -----------------------------------
quad_counts = ri1.groupby(['make','model','fuel_type_new','engine_cc_bin'])['co2_imputed'].transform('size')
mode_quad = (
    ri1[quad_counts >= 10]
      .groupby(['make','model','fuel_type_new','engine_cc_bin'])['co2_imputed']
      .agg(lambda s: s.mode().iat[0] if not s.mode().empty else np.nan)
)
mask_quad = ri1['co2_imputed'].isna() & quad_counts.ge(10)
ri1.loc[mask_quad, 'co2_imputed'] = (
    ri1.loc[mask_quad]
       .set_index(['make','model','fuel_type_new','engine_cc_bin'])
       .index.map(mode_quad)
)
print("After quadruple-mode, NaNs:", ri1['co2_imputed'].isna().sum())

# 3. Triple group mode imputation
# --------------------------------
ri1['co2_imputed'] = ri1['co2_imputed'].fillna(ri1['co2_emissions_clean'])
triple_counts = ri1.groupby(['make','model','fuel_type_new'])['co2_imputed'].transform('size')
mode_triple = (
    ri1[triple_counts >= 20]
      .groupby(['make','model','fuel_type_new'])['co2_imputed']
      .agg(lambda s: s.mode().iat[0] if not s.mode().empty else np.nan)
)
mask_triple = ri1['co2_imputed'].isna() & triple_counts.ge(20)
ri1.loc[mask_triple, 'co2_imputed'] = (
    ri1.loc[mask_triple]
       .set_index(['make','model','fuel_type_new'])
       .index.map(mode_triple)
)
print("After triple-mode, NaNs:", ri1['co2_imputed'].isna().sum())

# 4. Triple group forward/backward fill
# --------------------------------------
ri1 = ri1.sort_values(['make','model','fuel_type_new'])
ri1['co2_imputed'] = (
    ri1
      .groupby(['make','model','fuel_type_new'])['co2_imputed']
      .transform(lambda s: s.ffill().bfill())
)
print("After ffill/bfill, NaNs:", ri1['co2_imputed'].isna().sum())

# 5. Final fallback: global median
# ---------------------------------
global_med = ri1['co2_imputed'].median()
ri1['co2_imputed'] = ri1['co2_imputed'].fillna(global_med)
print("After global median, NaNs:", ri1['co2_imputed'].isna().sum())

# 6. (Optional) Visualize resulting distribution
#plt.figure(figsize=(10,4))
#ri1['co2_imputed'].hist(bins=30)
#plt.title("Distribution of Imputed CO2 Emissions")
#plt.xlabel("CO2 Emissions (g/km)")
#plt.ylabel("Count")
#plt.tight_layout()
#plt.show()
```
