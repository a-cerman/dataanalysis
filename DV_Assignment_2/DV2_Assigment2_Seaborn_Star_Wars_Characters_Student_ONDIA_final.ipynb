It provides a high-level interface for drawing **fascinating, attractive and informative graphics**. Moreover, Seaborn is a library for making **statistical graphics** in Python. It builds on top of matplotlib and integrates closely with pandas data structures.

Seaborn helps you explore and understand your data. Its plotting functions operate on dataframes and arrays containing whole datasets and internally perform the necessary semantic mapping and statistical aggregation to produce informative plots. Its dataset-oriented, declarative API lets you focus on what the different elements of your plots mean, rather than on the details of how to draw them

## <p style="background-color:#9d4f8c; font-family:newtimeroman; color:#FFF9ED; font-size:150%; text-align:center; border-radius:10px 10px;">"TASK_1 Plotting the Star Wars characters by genders"</p>

```python
gender=df1.gender.value_counts()
gender
```

```python
# gender
```

**This can now be plotted much more concisely.**<br>
Axes.bar_label() automatically labels bars [since matplotlib 3.4](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.bar_label.html)

```python
# with seaborn barplot
plt.figure()
ax1=sns.barplot(gender,palette='RdBu')
for i in ax1.containers:
    ax1.bar_label(i,)
plt.xticks(rotation=45)
plt.show()
```

```python
# with seaborn countplot
sns.countplot(x=df1.gender,order=gender.index, palette='RdBu')
for i in ax1.containers:
    ax1.bar_label(i,)
plt.xticks(rotation=45)
plt.show()
```

## <p style="background-color:#9d4f8c; font-family:newtimeroman; color:#FFF9ED; font-size:150%; text-align:center; border-radius:10px 10px;">TASK_3 - Plotting the MASS under 3 groups: "High_mass", "Normal_mass", "Low_mass" </p>


<a href="#toc" class="btn btn-primary btn-sm" role="button" aria-pressed="true"
style="color:blue; background-color:#dfa8e4" data-toggle="popover">Content</a>

```python
#mass_cat
bin=[0,50,99,int(df1.mass.max())]
labels=['Low_mass','Normal_mass','High_mass']
df1['mass_cat']=pd.cut(df1.mass, bins=bin,labels=labels,right=True)
df1['mass_cat'].value_counts(dropna=False)
```

```python
# with seaborn barplot
massc=df1['mass_cat'].value_counts()
ax=sns.barplot(massc,palette='RdBu',order=massc.index)
ax.set_title('Mass Groups',color='b',fontsize=15)
```

**Seaborn color palettes**

sns.color_palette("bone", n_colors=20)


https://seaborn.pydata.org/tutorial/color_palettes.html
https://www.practicalpythonfordatascience.com/ap_seaborn_palette

```python
# with seaborn countplot
plt.figure(figsize=(8,5))
ax=sns.countplot(df1,x='mass_cat',palette='RdBu',order=massc.index,width=0.8)
ax.set_title('Mass Groups',color='b',fontsize=15)
plt.xticks(rotation=45)
#sns.set_style(rc={'ax.facecolor': 'white', 'figure.facecolor': 'white'})
ax.set_facecolor('lightblue')
sns
for i in ax.containers:
    ax.bar_label(i,color='r')
```

```python
#species_freq
x4=df1.species.value_counts()[:5]
x4
```

https://www.statology.org/seaborn-barplot-color/

```python
# with seaborn Barplot

ax=sns.barplot(x4,palette='RdBu')
ax.set_title('Species Top 5',color='c',fontsize=15)
ax.set(ylabel='')
for i in ax.containers:
    ax.bar_label(i)
```

```python
# With Seaborn Countplot
ax=sns.countplot(df1,x='species',order=x4.index,palette='RdBu')
ax.set_title('Species Top 5',color='c',fontsize=15)
ax.set(ylabel='')
for i in ax.containers:
    ax.bar_label(i,fmt='%d.0',padding=-14)
```
